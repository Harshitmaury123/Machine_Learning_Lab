```python
import pandas as pd
df=pd.read_csv('advertising.csv')
df.head()
```

```python
x=df.drop('Sales',axis=1)  #(200,3)
y=df['Sales'] #(200,1)
y=y.T  #(1,200)
```

```python
import numpy as np
def Linear_Regression(x,y):
  weights=np.zeros((1,x.shape[1])) #(1,3)
  bias=0
  learning_rate=1e-7
  iterations=1000
  for i in range(iterations):
   y_pred=np.dot(weights,x.T)+bias #(1,3)*(3,200) -> (1,200)
   loss=np.mean((y-y_pred)**2)   #(1,)
   print('-'*100)
   print(loss)
   print('-'*100)
   dw=2*(y_pred-y).dot(x)
   db=2*np.mean(y_pred-y)
   weights=weights-learning_rate*dw
   bias=bias-learning_rate*db
  return weights,bias
```

```python
y=np.array(y)
x=np.array(x)
w,b=Linear_Regression(x,y)

print(b)
print(w)
```

```python
def predict(x):
  y_pred=np.dot(w,x.T)+b
  return y_pred

a=np.array([230.1,37.8,69.2])

print(predict(a))
```
